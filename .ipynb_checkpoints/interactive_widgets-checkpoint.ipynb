Let us view both these signals

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.io.wavfile import read, write
import scipy
```

```python
ac = read("/home/nipun/out.wav")[1]
```

```python
plt.plot(ac)
plt.xlabel("Sample #")
plt.ylabel("Amp")
plt.title("Noisy AC!");
```

```python
rec = read("/home/nipun/nipun.wav")[1]
```

```python
plt.plot(ac)
plt.xlabel("Sample #")
plt.ylabel("Amp")
plt.title("Recording");
```

Can't see much in the time domain. Let us make the spectrograms for these two signals.

```python
plt.specgram(ac);
plt.colorbar();
plt.xlabel("Sample #")
plt.ylabel("Normalized frequency");
plt.title("AC");
```

```python
plt.specgram(rec);
plt.colorbar();
plt.xlabel("Sample #")
plt.ylabel("Normalized frequency");
plt.title("Recording");
```

Let us write a small function to do low pass or high pass filtering

```python
from scipy import signal
```

```python
def bpf(N, f, lpf = True):
    h = signal.firwin(N, f, pass_zero=lpf)
    y= signal.lfilter( h, 1.0, rec)
    plt.specgram(y);
    plt.colorbar();
      
```

```python
bpf(1, 0.9, True)
```

```python
bpf(100, 0.9, True)
```

```python
bpf(101, 0.3, False)
```

```python
interactive_bpf = interactive(bpf, N = (1, 101, 2), f = (0.0, 0.9, 0.1))
```
